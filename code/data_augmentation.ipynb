```python
import smogn
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('./TRAIN.csv')
data_smogn = smogn.smoter(
    data = data,
    y = 'z'
)
```

```python
sns.kdeplot(data['z'], label='Original')
sns.kdeplot(data_smogn['z'], label='Augmented')
```

```python
data_smogn.to_csv('./TRAIN_smogn', header=False, index=False)
```
